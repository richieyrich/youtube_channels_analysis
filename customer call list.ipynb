```python
import pandas as pd
import numpy as np

# read data
df = pd.read_excel('Customer Call List.xlsx')
df.head(9)
```

```python
df.info()
```

```python
# finding duplicates
d = df[df.duplicated()]

# remove duplicates
df = df.drop_duplicates()
df.head(9)
```

```python
# chk null values
df.isnull().sum()

# remove null values in phne numbr column
df = df.dropna(subset=['Phone_Number'])
#df['Phone_Number'] = df['Phone_Number'].dropna()
df.isnull().sum()
```

```python
# clean fname from extra characters
df['Last_Name'] = df['Last_Name'].str.strip('\/._')
df
```

```python
# clean the phn num column
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
# df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]', '', regex=True)
df

df
```

```python
# format the phne num
df['Phone_Number'] = df['Phone_Number'].astype(str)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[:3]+'-'+ x[3:6]+'-'+ x[6:])
df
```

```python
# remove non number rows from phne num
# df.info()
dfa = df[df['Phone_Number'].str.contains('[^0-9-]')].index
df = df.drop(dfa).reset_index(drop=True)
df
```
